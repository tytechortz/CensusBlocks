```python
import geopandas as gpd
import pandas as pd
```

```python
geo_data = gpd.read_file('Census_Blocks_2020_SHAPE_WGS/Census_Blocks_2020_WGS.shp')
geo_data
```

```python
df = pd.read_csv('2020Census_BlockPop.csv')
df
```

```python
df['Total'] = df['Total'].shift(-1)
df
```

```python
df = df.dropna()
df
```

```python
df['Tract'] = df['Label (Grouping)'].str.extract(r'Tract (.*?),', expand=False )
df
```

```python
df['Tract'] = df['Tract'].astype(float)
# df['Tract'] = df['Tract']*100
df
```

```python
df['Tract'] = 100 * df['Tract']
df
```

```python
df['Tract'] = df['Tract'].round(0)
df
```

```python
df['Tract'] = df['Tract'].astype(float).astype(int).astype(str)
df
```

```python
df['Block Group'] = df['Label (Grouping)'].str.extract(r'Group (.*?),', expand=False )
# pd.set_option('display.max_rows', None)
df
```

```python
df['Block'] = df['Label (Grouping)'].str.extract(r'Block (.*?),', expand=False )
# pd.set_option('display.max_rows', None)
df
```

```python
df['Tract'] = df['Tract'].str.rjust(6, '0')
df
```

```python
df['GEOID'] = '08005' + df['Tract'] + df['Block']
df
```

```python
df.to_csv('BlockPop.csv')
```
