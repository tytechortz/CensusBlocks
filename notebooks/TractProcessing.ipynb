```python
import geopandas as gpd
import pandas as pd

df = pd.read_csv('2020Census_TractPop.csv')
df
```

```python
df['Total'] = df['Total'].shift(-1)
df
```

```python
df = df.dropna()
df
```

```python
geo_data = gpd.read_file('2020_CT/ArapahoeCT.shp')
geo_data['FIPS']
```

```python
import re

df['Tract'] = df['Label (Grouping)'].str.extract(r'Tract (.*?);', expand=False )
df
```

```python
df['Tract'] = df['Tract'].astype(float)
# df['Tract'] = df['Tract']*100
df
```

```python
df['Tract'] = 100 * df['Tract']
df
```

```python
df['Tract'] = df['Tract'].round(0)
df
```

```python
df['Tract'] = df['Tract'].astype(float).astype(int).astype(str)
df
```

```python
df['Tract'] = df['Tract'].str.rjust(6, '0')
df
```

```python
df['GEOID'] = '08005' + df['Tract']
df
```

```python
df.to_csv('TractPop.csv')
```
