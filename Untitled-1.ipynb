```python
df = df[df['Label (Grouping)'].str.contains('Estimate')==False]
df
```

```python
import re

df['Tract'] = df['Label (Grouping)'].str.extract(r'Tract (.*?),', expand=False )
df
```

```python
df['Tract'] = df['Tract'].str.replace('.', '')
df
```

```python
df['Block Group'] = df['Label (Grouping)'].str.extract(r'Group (.*?),', expand=False )
df
```

```python
df['Tract'] = df['Tract'].apply(lambda x: '{0:0>6}'.format(x)) 
df
```

```python
df['GEOID'] = '08005' + df['Tract'] + df['Block Group']
df
```
